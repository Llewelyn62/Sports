```python
def Massey_Rank(A, Results):
    ''' A is the game matrix with diagonals = 1, Results is the 
    score difference for each game'''
    A = np.array(A)
    Results = np.array(Results)
    #Now multiply both sides by A.T
    M = A.T*A
    R = A.T*Results
    #Replace Row m with ones and row m in Results with 0
    #m = M.cols
    M[-1:] = 1; R[-1] = 0
    #lstsq returns the solution, rank and singular values
    Solution = np.linalg.lstsq(M,R)
    return M, R, Solution[0]
```

### 1. Analysis of team strength using the Massey system. 

Essentially, the sum of goal differences is the dependent variable. It turns out to be a fairly powerful metric over time. Each team plays a combination of other teams, this combination forms the matrix of who plays who. This then allows the goal difference to be keyed to specific teams. A least squares projection allows estimates of the weight each team carries, irrespective of who they play. This can then be used to estimate/compare two teams against each other and predict which is stronger.

### 2. Bayesian analysis to predict goal differences across teams

The following analysis uses a combination of the Massey dependent variable (sum of goal differences) and the Bayesian idea that there is a level of certainty about the range of what these goal difference sums ought to be. So, for example, RBHS won three games and accumulated 15 points more than the teams they played. What is the distribution likely to be if these three games could be repeated a sufficiently large number of times and, more usefully, what is the goal difference range likely to range between. The mean and range of scores (between 2.5% and 97.5%) represents a 95% confidence that the goal difference would be in that range. Compare (subtract) this from another team's difference matrix and find the likely range of outcomes. Ultimately, this allows some prediction of probability of winning over losing, and by how much. 

```python
#Collect the data from the dataframe. Note the format
stan_data = {
    'N' : s15df.shape[0],
    s15df.columns[0]:   list(s15df.iloc[:,0].astype(int)),
    s15df.columns[1]:   list(s15df.iloc[:,1].astype(int)),
    s15df.columns[2]:   list(s15df.iloc[:,2].astype(int)),    
    s15df.columns[3]:   list(s15df.iloc[:,3].astype(int)),
    s15df.columns[4]:   list(s15df.iloc[:,4].astype(int)),
    s15df.columns[5]:   list(s15df.iloc[:,5].astype(int)),
    s15df.columns[6]:   list(s15df.iloc[:,6].astype(int)),
    s15df.columns[7]:   list(s15df.iloc[:,7].astype(int))
}
stan_data,s15df.shape[0]
```

```python
#The Stan Code Normal distribution
#+++++++++++++++++++++++++++++++++++++++++
stan_code_n = '''
data {
    int N;
    real JPC[N];
    real RBHS[N];
    real Ruamata[N];
    real TNT[N];
    real Tauhara[N];
    real Trident[N];
    real WHHS[N];
    real RLHS[N];
}
transformed data {}
parameters {
    real muJPC;
    real<lower=0, upper=10> sigmaJPC;
    real muRBHS;
    real<lower=0, upper=10> sigmaRBHS;
    real muRuamata;
    real<lower=0, upper=10> sigmaRuamata;
    real muTNT;
    real<lower=0, upper=10> sigmaTNT;
    real muTauhara;
    real<lower=0, upper=10> sigmaTauhara;
    real muTrident;
    real<lower=0, upper=10> sigmaTrident;
    real muWHHS;
    real<lower=0, upper=10> sigmaWHHS;
    real muRLHS;
    real<lower=0, upper=10> sigmaRLHS;
}
transformed parameters {}
model {
    JPC ~ normal(muJPC, sigmaJPC);
    RBHS ~ normal(muRBHS, sigmaRBHS);
    Ruamata ~ normal(muRuamata, sigmaRuamata);
    TNT ~ normal(muTNT, sigmaTNT);
    Tauhara ~ normal(muTauhara, sigmaTauhara);
    Trident ~ normal(muTrident, sigmaTrident);
    WHHS ~ normal(muWHHS, sigmaWHHS);
    RLHS ~ normal(muRLHS, sigmaRLHS);

}
generated quantities {
    real RBHS_TRI;
    real RBHS_Rua;
    real RBHS_TNT;
    Cohen_d_RBHS_TRI = (muRBHS-muTauhara)/sqrt(sigmaRBHS^2 + sigmaTauhara^2);
    Cohen_d_RBHS_Rua = (muRBHS-muRuamata)/sqrt(sigmaRBHS^2 + sigmaRuamata^2);
    Cohen_d_RBHS_TNT = (muRBHS-muTNT)/sqrt(sigmaRBHS^2 + sigmaTNT^2);

}
'''
```

```python
sm = ps.StanModel(model_code=stan_code_n)
fit = sm.sampling(data=stan_data)
```

### 3. Using Bayesian Poisson Distribution to predict scoring rates

An alternative to looking at score difference is to use actual scores for each team to predict scoring rate. 

https://www.pinnacle.com/en/betting-articles/soccer/how-to-calculate-poisson-distribution#height

```python
PDF('http://sportsrunner.net/SR_RRH/filespdf/SB_Results.pdf?d=1501206919704',size=(400,400))
```

```python
stan_data_p = {
    'N' : s15df.shape[0],
    'JPC' : [6,0,1,2,7,7],
    'RBHS': [2, 9, 6,2,3,3],
    'Ruamata': [1,3,1,3,0,0],
    'TNT': [7, 12, 16,1,2,2],
    'Tauhara': [6, 1, 2,1,2,2],
    'Trident': [1, 5, 7,1,0,0],
    'WHHS': [3,1,1,3,1,1],
    'RLHS': [0,0,0,0,0,0]
}
#The Stan Code Poisson Distribution
#+++++++++++++++++++++++++++++++++++++++++
stan_code_p = '''
data {
    int N;
    int JPC[N];
    int RBHS[N];
    int Ruamata[N];
    int TNT[N];
    int Tauhara[N];
    int Trident[N];
    int WHHS[N];
    int RLHS[N];
}
transformed data {}
parameters {
    real <lower=0> lamJPC;
    real <lower=0> lamRBHS;
    real <lower=0> lamRuamata;
    real <lower=0> lamTNT;
    real <lower=0> lamTauhara;
    real <lower=0> lamTrident;
    real <lower=0> lamWHHS;
    real <lower=0> lamRLHS;
}
transformed parameters {}
model {
    JPC ~ poisson(lamJPC);
    RBHS ~ poisson(lamRBHS);
    Ruamata ~ poisson(lamRuamata);
    TNT ~ poisson(lamTNT);
    Tauhara ~ poisson(lamTauhara);
    Trident ~ poisson(lamTrident);
    WHHS ~ poisson(lamWHHS);
    RLHS ~ poisson(lamRLHS);
}
generated quantities {
    int RBHS_p;
    int TRI_p;
    int TNT_p;
    int Rua_p;
    RBHS_p = poisson_rng(lamRBHS);
    TRI_p = poisson_rng(lamTrident);
    TNT_p = poisson_rng(lamTNT);
    Rua_p = poisson_rng(lamRuamata);
}
'''
```

```python
sm2 = ps.StanModel(model_code=stan_code_p)
fit_poisson = sm2.sampling( data=stan_data_p, iter=1000, chains=4)
```

```python
print(fit_poisson)
```

```python
fit_poisson.traceplot(['RBHS_p', 'TNT_p', 'TRI_p', 'Rua_p'])
```

```python
trace_p = fit_poisson.extract()
total_samples = np.count_nonzero(trace_p['RBHS_p'])
table1RBHS = np.array([np.count_nonzero(trace_p['RBHS_p'][:] > i) for i in range(16)])/total_samples
total_samples = np.count_nonzero(trace_p['TNT_p'])
table1TNT = np.array([np.count_nonzero(trace_p['TNT_p'][:] > i) for i in range(16)])/total_samples
total_samples = np.count_nonzero(trace_p['TRI_p'])
table1TRI = np.array([np.count_nonzero(trace_p['TRI_p'][:] > i) for i in range(16)])/total_samples
total_samples = np.count_nonzero(trace_p['Rua_p'])
table1Rua = np.array([np.count_nonzero(trace_p['Rua_p'][:] > i) for i in range(16)])/total_samples
#Plot the cumulative results
fig,ax1 = plt.subplots(figsize=(12,8))
ax1.plot(table1RBHS,color='Red',alpha=.5)
ax1.plot(table1TNT,color='Black',alpha=.5)
ax1.plot(table1TRI,color='Blue',alpha=.5)
ax1.plot(table1Rua,color='Brown',alpha=.5)
ax1.legend(['RBHS','TNT', 'Trident','Ruamata'])
ax1.set_title('Poisson Distribution of cumulative scoring rates',
             size=16)
ax1.set_xlabel('Rate of scoring (goals per game)')
ax1.set_ylabel('Probability of a particular rate')
```
